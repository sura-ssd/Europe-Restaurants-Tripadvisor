```python
import pandas as pd
import numpy as np
```

```python
#reading CSV file
tripadvisor_df = pd.read_csv('tripadvisor_european_restaurants.csv')
```

```python

# Extract reviews data
tripadvisor_df.head(5)
```

```python
#dropping the irrelevant columns
columns_to_drop = ['restaurant_link',
'claimed',
'popularity_detailed',
'popularity_generic',
'top_tags',
'price_range',
'meals',
'special_diets',
'vegetarian_friendly',
'vegan_options',
'gluten_free',
'original_open_hours',
'open_days_per_week',
'open_hours_per_week',
'working_shifts_per_week',
'food',
'service',
'value',
'atmosphere',
'keywords',]
tripadvisor_df.drop(columns=columns_to_drop, inplace=True)
```

```python
columns_to_drop =['reviews_count_in_default_language', 'default_language']
tripadvisor_df.drop(columns=columns_to_drop, inplace=True)
```

```python
tripadvisor_df.columns
```

```python
#Dropping duplicate cells
tripadvisor_df = tripadvisor_df.drop_duplicates()
```

```python
tripadvisor_df.head(5)
```

```python
# Handle missing values
tripadvisor_df = tripadvisor_df.fillna(-1).dropna()
tripadvisor_df.dropna(inplace=True)  # drop rows with NaNs in any column
```

```python
tripadvisor_df.head(5)
```

```python
# Import statement to work with SQLAlchemy database connection 
from sqlalchemy import create_engine
```

```python
# Create engine to connect to database
engine = create_engine('postgresql://postgres:user@localhost:5432/RestaurantReview')
```

```python
# Convert DataFrame to SQL table
tripadvisor_df.to_sql("RestaurantReview", engine, if_exists="replace", index=False)
```
